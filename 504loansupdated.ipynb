```python
# Getting the total gross approval amount per bank, by year

GrossApprovalCompare = Santander_vs_TopBanks.groupby(['ApprovalFiscalYear','ThirdPartyLender_Name'])['GrossApproval'].sum().unstack()
```

```python
GrossApprovalCompare
```

```python
import seaborn as sns
from matplotlib import pyplot as plt


GrossApprovalCompare.plot(kind='bar', stacked=True, figsize=(20,10))


plt.title('Gross Distributions per Bank By Year(100 Millions)')
plt.xticks(rotation=0, ha='center')
```

```python
NAICS_Compare=Santander_vs_TopBanks.groupby(['ApprovalFiscalYear','Supersector'])['GrossApproval'].sum().unstack().fillna(0)
```

```python
NAICS_Compare
```

```python
NAICS_Compare.plot(kind='bar', stacked=True, figsize=(20,10))


plt.title('Gross Distributions per Sector By Year for Top Banks')
plt.xticks(rotation=0, ha='center')
```

```python
#Number of distributions of top banks(and Santander) by year

Num_of_distributions_per_year =  Santander_vs_TopBanks.groupby(['ApprovalFiscalYear','ThirdPartyLender_Name']).size().unstack().fillna(0)
```

```python
Num_of_distributions_per_year
```

```python
Num_of_distributions_per_year.plot(kind='bar', stacked=True, figsize=(20,10))


plt.title('Number of Distributions Each Year Per Bank')
plt.xticks(rotation=0, ha='center')
```

```python
Num_of_distributions_per_year_percent_change = Num_of_distributions_per_year.pct_change()
```

```python
Num_of_distributions_per_year_percent_change
```

```python
Num_of_distributions_per_year_percent_change.plot(figsize=(20,10), xticks = (2017,2018,2019,2020,2021))

plt.title('Number of Distributions Each Year Per Bank PCT Change')
plt.xticks(rotation=0, ha='center')
```

```python
# In 2020, Eastern Bank created a impact loan fund for small business in need during Covid, which isd why there is a steady rise of distributions.
```
